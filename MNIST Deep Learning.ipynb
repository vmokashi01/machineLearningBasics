```python
import numpy as np
import pandas as pd
from tensorflow import keras
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, Conv2D, Flatten, Dropout
from sklearn.model_selection import train_test_split
```

```python
img_rows = 28
img_cols = 28
classes = 10
```

```python
def prep(raw):
    out_y = keras.utils.to_categorical(raw.label, classes)
    num_images = raw.shape[0]
    x_as_array = raw.values[:,1:]
    x_shaped_arr = x_as_array.reshape(num_images, img_rows, img_cols, 1)
    out_x = x_shaped_arr/255
    return out_x, out_y
```

```python
train_file = "../input/digit-recognizer/train.csv"
raw_data = pd.read_csv(train_file)
```

```python
x, y = prep(raw_data)
```

```python
model = Sequential()
model.add(Conv2D(20, kernel_size = 3, activation ='relu', input_shape = (img_rows, img_cols, 1)))
model.add(Conv2D(12, kernel_size = 3, activation ='relu'))
model.add(Conv2D(12, kernel_size = 3, activation ='relu'))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(classes, activation='softmax'))
```

```python
model.compile(loss=keras.losses.categorical_crossentropy, optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x,y,batch_size = 128, epochs = 4, validation_split = 0.2)
```

```python
test_file = "../input/digit-recognizer/test.csv"
test_data = pd.read_csv(test_file)
```

```python
def prep_test(input):
    num_images = input.shape[0]
    x_as_array = input.values[:,:]
    x_shaped_arr = x_as_array.reshape(num_images, img_rows, img_cols, 1)
    out_x = x_shaped_arr/255
    return out_x
```

```python
x_test = prep_test(test_data)
```

```python
predictions = model.predict(x_test, batch_size = 128, verbose = 1)
```

```python
predictions = np.argmax(predictions,axis = 1)

predictions = pd.Series(predictions,name="Label")
```

```python
submission = pd.concat([pd.Series(range(1,28001),name = "ImageId"),predictions],axis = 1)
```
